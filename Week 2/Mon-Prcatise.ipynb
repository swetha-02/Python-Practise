```python
import requests
from bs4 import BeautifulSoup
```

```python
url="https://www.ebay.com/sch/i.html?_from=R40&_trksid=p4432023.m570.l1313&_nkw=mobiles&_sacat=0"
header={'User-Agent':"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/123.0.0.0 Safari/537.36"}
data=requests.get(url,header)
```

```python
data.content
```

```python
soup=BeautifulSoup(data.content,'html5lib')
```

```python
soup
```

```python
mobile=soup.findAll('div',attrs={'class':"s-item__title"})[1:51]
```

```python
mobile
```

```python
mobile_s=[]
```

```python
for i in mobile:
        mobile_s.append(i.span.get_text())
```

```python
mobile_s
```

```python
moblile_price=soup.findAll('span',attrs={'class':"s-item__price"})[1:51]
```

```python
moblile_price
```

```python
mobile_p=[]
```

```python
for i in moblile_price:
    mobile_p.append(i.get_text())
```

```python
mobile_p
```

```python
import pandas as pd
```

```python
df_mobiles=pd.DataFrame(mobile_s,columns=["Mobile Phones"])
```

```python
df_mobiles
```

```python
df_mobprice=pd.DataFrame(mobile_p,columns=["Price"])
```

```python
df_mobprice
```

```python
df=[df_mobiles,df_mobprice]
df_mobile_price=pd.concat(df,axis=1)
```

```python
df_mobile_price
```

```python
df_mobile_price.to_csv("Mobile details.csv")
```

```python
df_mobile_price.count()
```
